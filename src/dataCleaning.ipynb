```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
```

```python
tweets = (open("./../Dataset/trial/us_trial.text", "r").readlines())
```

```python
allTweets = tweets
allTweets
```

```python
import re
import string
from tqdm import tqdm
```

```python
for sent_ind in tqdm(range(len(allTweets))):
    sent = allTweets[sent_ind]
    allTweets[sent_ind] = re.sub(r'@user','',sent)
```

```python
for ind in tqdm(range(len(allTweets))):
    allTweets[ind]  = "".join([char.lower() for char in allTweets[ind] if char not in string.punctuation])
    allTweets[ind]  = "".join([char.lower() for char in allTweets[ind] if char not in ['…','\n']])
    allTweets[ind] = re.sub('[0-9]+', '', allTweets[ind])
    allTweets[ind] = re.sub('\s+', ' ', allTweets[ind])
```

```python
import nltk
```

```python
stopword = nltk.corpus.stopwords.words('english')
stopword += ['yr', 'year', 'woman', 'man', 'girl','boy','one', 'two', 'sixteen', 'yearold', 'fu', 'weeks', 'week',
        'treatment', 'associated', 'patients', 'may','day', 'case','old']
```

```python
from nltk.tokenize import word_tokenize
```

```python
for i in tqdm(range(len(allTweets))):
    allTweets[i] = word_tokenize(allTweets[i])
```

```python
for i in tqdm(range(len(allTweets))):
    allTweets[i] = [word for word in allTweets[i] if word not in stopword]
```

```python
from textblob import TextBlob
```

```python
for i in tqdm(range(50):
    allTweets[i] = [str(TextBlob(word).correct()) for word in allTweets[i]]
```

```python
for i in tqdm(range(len(allTweets))):
    allTweets[i] = [word for word in allTweets[i] if word not in stopword]
```

```python
from nltk.stem import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
for i in tqdm(range(len(allTweets))):
    allTweets[i] = [ps.stem(word) for word in allTweets[i]]
```
