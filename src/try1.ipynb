```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
```

```python
tweets = (open("./../Dataset/trial/us_trial.text", "r").readlines())
emoji = (open("./../Dataset/trial/us_trial.labels", "r").readlines())
```

```python
allTweets = pd.DataFrame(tweets, columns=['tweets'])
allEmoji = pd.DataFrame(emoji, columns=['emoji'])
```

```python
from sklearn.model_selection import train_test_split
```

```python
trainTweets, testTweets, trainEmoji, testEmoji = train_test_split(allTweets, allEmoji, test_size=0.2, random_state=0)
```

```python
trainTweets.reset_index(drop=True, inplace=True)
testTweets.reset_index(drop=True, inplace=True)
trainEmoji.reset_index(drop=True, inplace=True)
testEmoji.reset_index(drop=True, inplace=True)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
```

```python
count_vect = CountVectorizer()
term_doc_matrix = count_vect.fit_transform(trainTweets['tweets'])
```

```python
tfidf_transformer = TfidfTransformer()
normalized_tdm = tfidf_transformer.fit_transform(term_doc_matrix)
```

```python
classifier = MLPClassifier().fit(normalized_tdm, trainEmoji['emoji'])
```

```python
predictions = classifier.predict(normalized_tdm)
print(accuracy_score(predictions,trainEmoji))
```
